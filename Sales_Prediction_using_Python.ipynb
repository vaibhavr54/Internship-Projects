### **_Exploratory Data Analysis_**

#### *Import Required Libraries*

```python
import numpy as np  #linear algebra
import pandas as pd  #data processing
```

```python
df = pd.read_csv("advertising.csv")
```

```python
x = df.iloc[:, 0:-1]
```

```python
y = df.iloc[:,-1]
```

```python
y
```

#### *Train Test Split*

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
x_train
```

```python
x_test
```

```python
x_train=x_train.astype(int)
y_train=y_train.astype(int)
x_test=x_test.astype(int)
y_test=y_test.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
x_train_scaled=Sc.fit_transform(x_train)
x_test_scaled=Sc.fit_transform(x_test)
```

#### *Applying Linear Regression*

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test_scaled)
```

#### *Evaluate the performance of a Linear Regerssion Model*

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

#### *Analyzing the data by scatter plot*

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(y_test,y_pred,c='g')
```

#### **_The scatter plot visualizes the relationship between the actual values (y_test) and the predicted values (y_pred) from our machine learning model. It indicates a positive correlation, where an increase in actual values corresponds to an increase in predicted values. The points closely align along the diagonal line, suggesting that the model's predictions are fairly accurate. Despite some deviations, the majority of the points cluster around the diagonal, demonstrating the model's consistency and overall good performance in predicting the target variable._**
